# EDA

```python
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt


sn.countplot(df['Firearms'])  #checking distribution of the column
plt.title("Firearms")
plt.show()
```

Shows that majority of the firearms cases are from Uttar Pradesh 

```python
sn.countplot(df['Licenced Firearms'])  #checking distribution of the column
plt.title("Licenced Firearms")
plt.show()
```

Shows that majority of the Licenced firearms cases are from Uttar Pradesh 

```python
sn.countplot(df['UnLicenced Firearms'])  #checking distribution of the column
plt.title("UnLicenced Firearms")
plt.show()
```

Shows that majority of the Licenced firearms cases are from Uttar Pradesh


```python
#ploting scatterplot to check relationship bw "size" and "Weight"
plt.scatter(df["Firearms"],df["UnLicenced Firearms"])
plt.title("Scatter Plot")
plt.xlabel("Firearms")
plt.ylabel("UnLicenced Firearms")
plt.show()
```

There is a slight linear relation ship between Firearms and UnLicenced firearms 

```python
#Checking corelation using heatmap

corr=df.corr()  # df.corr function provides the corelation values of each column 
sn.heatmap(corr,annot=True) #using heatmap we are ploting correlation values obtained by df.corr function.......
#annot=True to put the numbers on the blocks of thr graph 
plt.show()
```

the heatmap shows that the dataaset is strongly corelated
